### Линейная регрессия

Требования:

1. Python >= 3.6
2. numpy 1.13.1
3. matplotlib 2.0.2

```python
import numpy as np
import matplotlib.pyplot as plt
```

Создадим небольшие данные, которые будем использовать

```python
data = np.linspace(1.0, 20.0, num=80)
target = np.linspace(3., -5., num=80) + np.random.normal(loc=0, scale=1, size=80)
```

```python
plt.plot(data, target, 'ro')
plt.xlabel('x')
plt.ylabel('target')
plt.show()
```

Как и на лекциях, мы подразумеваем, что $p(y|x, w, b) = \mathcal{N}(y | x*w + b, \sigma^2)$

Имея цель найти максимум правдоподобия, мы говорим, что $w, b = argmin_{w, b}\big[(Xw + b - y)^2\big] = argmin_{w, b} Loss$, где $X$ – матрица $N \times k$, $N$ – кол-во наблюдений, $k$ = размерность каждого наблюдения (в нашем случае $k$ = 1).

Заметим, что $xw + b$ может быть записано как $[x|1]\theta$ (то есть мы дополнили каждое наблюдение единицей справа, чтобы таким образом к скалярному умножение $x$ и $w$ добавить $b$, который никак не зависит от $x$)

Найдем closed-form solution для $\theta$ как $\theta = (X^{\top}X)^{-1}X^{\top}y$

```python
expanded_data = np.concatenate([data.reshape(-1,1), np.ones(80).reshape(-1,1)], axis=1)
```

```python
from numpy.linalg import inv
closed_form_solve = inv(expanded_data.T @ expanded_data) @ expanded_data.T @ target
```

Посмотрим, что нам будет выдавать такое решение

```python
prediction = expanded_data @ closed_form_solve

plt.plot(data, target, 'ro')
plt.plot(data, prediction, 'g')

plt.xlabel('x')
plt.ylabel('target')

plt.show()
```

Мы видим, что closed-form solution выдало нам такие параметры линии, что она старается минимизировать расстояние до каждого из наших наблюдений

Однако как мы и говорили раньше, closed-form solution обладает рядом минусов, с которыми мы не готовы мириться.

Попробуем получить параметры модели численно максимизируя правдоподобие

$\nabla_{\theta} Loss =2X^{\top} (X\theta - y)$

Тогда используя градиентный спуск мы можем получить

```python
theta = np.array([1., 1.])

num_iterations = 10000
learning_rate = 0.00002

for i in range(num_iterations):
    grad_loss = 2 * expanded_data.T @ (expanded_data @ theta - target)
    theta = theta - learning_rate * grad_loss
```

```python
prediction = expanded_data @ theta

plt.plot(data, target, 'ro')
plt.plot(data, prediction, 'g')

plt.xlabel('x')
plt.ylabel('target')

plt.show()
```

Как ни странно, мы получили примерно то же самое

```python
print(theta)
print(closed_form_solve)
```

Попробуем добиться того же самого при помощи стохастического градиентного спуска

```python
s_theta = np.array([1., 1.])

num_iterations = 10000
learning_rate = 0.0001
batch_size = 10

for i in range(num_iterations):
    indexes = np.random.randint(80, size=batch_size)
    minibatch_data = np.stack([expanded_data[i] for i in indexes])
    minibatch_target = np.stack([target[i] for i in indexes])

    grad_loss = 2 * minibatch_data.T @ (minibatch_data @ s_theta - minibatch_target)
    s_theta = s_theta - learning_rate * grad_loss
```

```python
print(s_theta)
```

```python
prediction = expanded_data @ s_theta

plt.plot(data, target, 'ro')
plt.plot(data, prediction, 'g')

plt.xlabel('x')
plt.ylabel('target')

plt.show()
```

Обратите внимание, что нам пришлось менять значение learning_rate, поскольку значение градиента зависит от batch_size

Очевидным решением этой проблемы будет деление значения Loss функции на batch_size чтобы мы всегда находились в одинаковом масштабе

### Логистическая регрессия

```python
xs = np.random.multivariate_normal(np.array([1., 0]), np.array([[1, 0.6], [0.6, 1]]), 200)
plt.plot(xs[:, 0], xs[:, 1], 'bx')
plt.axis('equal')


os = np.random.multivariate_normal(np.array([2.5, -1]), np.array([[1, 0.6], [0.6, 1]]), 200)
plt.plot(os[:, 0], os[:, 1], 'ro')
plt.axis('equal')
plt.show()

data = np.concatenate([xs, os])
data = np.concatenate([data, np.ones(400).reshape(-1,1)], axis=1)
target = np.array([0] * 200 + [1] * 200)
```

Теперь наши данные – это матрица $300 \times 2$. 

+1 к размерности данных из-за нашей репараметризации = $300 \times 3$



Распределение вероятностей над двумя классами описывается при помощи распределения Бернулли 

$p(y) = p(y = 1)^y (1 - p(y =1))^{1 - y}$ где $y \in \{0, 1\}$

Мы можем задать параметр $p(y=1)$ как $f(x^{\top}\theta)$ ($f$ в данном случае необходима чтобы нормализовать и привести к удовлетворительному виду значение вероятности, т.к. мы знаем, что $p(y=1) \in [0; 1]$. Исходя из необходимых свойств, о которых мы еще поговорим, $f$ принимает вид $f = \frac{1}{1 + e^{-x}}$), тогда

$p(y|x, \theta) = f(x^{\top}\theta)^y (1 - f(x^{\top}\theta))^{1 - y}$

```python
import math

x = np.linspace(-10, 10, num=100)
y = 1. / (1 + np.exp(-x))

plt.plot(x, y, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Пытаясь найти максимум $p(y|X, \theta) = \prod_{(x_i, y_i)} \big(p(y_i|x_i, \theta)\big)$ относительно параметров $\theta$, мы можем рассматривать $ln \big[p(y|X, \theta)\big] = \sum_{(x_i, y_i)} \big[ln \big(p(y_i|x_i, \theta)\big)\big]$, 

т.к. $argmax_{\theta} \big(p(y|x, \theta)\big) = argmax_{\theta} \big(ln \big[p(y|x, \theta)\big]\big)$

$ln p(y|x, \theta) = y \big(ln(f(x^{\top}\theta))\big) + (1 - y) \big(ln(1 - f(x^{\top}\theta))\big)$

Чтобы найти минимум этой функции, нам всего лишь нужно рассмотреть $-ln p(y|X, \theta) = Loss$

$\nabla_{\theta}Loss = -\frac{1}{N}\sum_{(x_i, y_i)} \big((y_i - f(x_i^{\top}\theta))x_i\big)$

В матричной форме записи
$\nabla_{\theta} = - \frac{1}{N} X^{\top}(y - f(X\theta))$

Применим стохастический градиентный спуск для поиска параметров, которые разделяли бы наши данные

```python
s_theta = np.random.normal(0., 0.3, size=3)

num_iterations = 50000
learning_rate = 0.001
batch_size = 20

def sigmoid(x):
    return 1./(1 + np.exp(-x))

for i in range(num_iterations):
    indexes = np.random.randint(400, size=batch_size)
    
    minibatch_data = np.stack([data[i] for i in indexes])
    minibatch_target = np.stack([target[i] for i in indexes])

    grad_loss = - (1./batch_size) * minibatch_data.T @ (minibatch_target - sigmoid(minibatch_data @ s_theta))
    s_theta = s_theta - learning_rate * grad_loss
```

```python
s_theta
```

Полученные параметры определяют двухмерную плоскость. Т.к. мы в наших данных определили, что кружки – это единицы, а крестики – нули, то прямая $x_1 * \theta_1 + x_2 * \theta_2 + \theta_3 = 0$ определяет границу нашего выбора.

Т.е. все точки, которые находятся с той стороны прямой, с которой находится сгусток кружков, будут обладать свойством 

$x^{\top}\theta > 0$, i.e. $f(x^{\top}\theta) > 0.5$

```python
plt.xlabel('x_1')
plt.ylabel('x_2')

plt.plot(xs[:, 0], xs[:, 1], 'bx')
plt.axis('equal')

plt.plot(os[:, 0], os[:, 1], 'ro')
plt.axis('equal')

xs_line = np.linspace(-3.0, 6.0, num=80)
ys_line = np.array([(-x * s_theta[0] - s_theta[2]) / s_theta[1] for x in xs_line])

plt.plot(xs_line, ys_line, 'g')

plt.show()
```

Что важно, вероятностная основа машинного обучения позволяет нам рассуждать о том, насколько наше предсказание может быть верным. Очевидно, что если мы возьмем точку гду-то около $(2; 0)$, мы окажемся между нашими классами. 

$x^{\top}\theta$ будет примерно равно 0

```python
sigmoid(np.array([2., 0, 1]) @ s_theta)
```

А значит $p(y \in кружочки|x, \theta) = sigmoid(x^{\top}\theta) = 0.48$

Поэтому важно понимать, что если мы возьмем такую точку и отнесем ее к классу "Крестики", то мы это сдеаем только потому что формально вероятность крестиков больше, однако эта точка все еще с достаточно большим процентом вероятности может быть и кружочком

### Мультиклассовая классификация

```python
xs = np.random.multivariate_normal(np.array([4., 4]), np.array([[1, -0.6], [-0.6, 1]]), 200)
plt.plot(xs[:, 0], xs[:, 1], 'bx')
plt.axis('equal')

os = np.random.multivariate_normal(np.array([2.5, -1.5]), np.array([[1, 0.6], [0.6, 1]]), 200)
plt.plot(os[:, 0], os[:, 1], 'ro')
plt.axis('equal')

ss = np.random.multivariate_normal(np.array([-2.5, 1]), np.array([[1, 0.6], [0.6, 1]]), 200)
plt.plot(ss[:, 0], ss[:, 1], 'gs')
plt.axis('equal')

plt.show()

data = np.concatenate([xs, os, ss])
data = np.concatenate([data, np.ones(600).reshape(-1,1)], axis=1)
target = np.array([[1., 0., 0.]] * 200 + [[0., 1., 0]] * 200 + [[0., 0., 1.]] * 200)
```

Решение задачи мультиклассовой классификации не сильно отличается от бинарной классификации – распределения, лежащие в их основе, очень похожи.

Однако, поскольку у нас больше не два класса, а несколько, мы больше не можем определить только один параметр распределения, а второй получить из знания о том, что сумма вероятностей равна единице.

Естественным способом ввести такое распределение будет в случае когда $y$ может принять три состояния

$p(y) = \mu_1^{y_1}\mu_2^{y_2}\mu_3^{y_3}$
$y \in \{(1, 0, 0), (0, 1, 0), (0, 0, 1)\}$
Обозначая $\mu_i = p(y_i = 1)$

При этом добавляется очевидное условие $sum_{\mu_i}\big( \mu_i \big)= 1$

При этом, если мы захотим линейно обусловить вероятности некоторым $x$ и параметрами модели, то это будет означать, что

$\mu = f(x^{\top}\theta)$

$p(y|x, \theta) = \mu_1^{y_1}\mu_2^{y_2}\mu_3^{y_3}$

Т.к. $x$ – это вектор размерности $k$, $mu$ – это вектор размерности $m$ – кол-во классов, то $\theta$ должна быть матрицей $k \times m$

В свою очередь, функция $f$ должна брать вектор длины $m$ с элементами $\in (-\inf; \inf)$ и отображать его в вектор, сумма элементов которого равна единице.

Если мы подраузмеваем, что $x^{\top}\theta$ – это ненормализованные логарифмы вероятности, то чтобы получить вероятности, нам нужно взять от них экспоненту и нормализовать

$\mu_i = \frac{e^{x^{\top}\theta_i}}{\sum_j e^{(x^{\top}\theta_j)}}$

Это называется softmax функцией, и мы будем всегда ее использовать для многоклассовой классификации.
Можно показать, что сигмоида является частным случаем softmax, однако если вы решаете задачу двухклассовой классификации, то стоит использовать именно её, чтобы не плодить лишних параметров


Если мы возьмем логарифм от $p(y|x, \theta)$, то мы получим 

$ln(p(y|x, \theta)) = \sum_i \big(y_i * ln(Softmax(x^{top}\theta)_i)\big) = y^{\top}ln(Softmax(\theta^{\top}x))$

Так как $y$ – это вектор, у которого все элементы, кроме одного, равного единице, равны нулю, то мы просто "забираем" необходимые нам вероятности и пытаемся минимизировать Loss на них

Таким образом, чтобы максимизировать правдоподобие, мы должны минимизировать функцию

$Loss = -\frac{1}{N} \sum_{i} ln\big(Softmax(X\theta)\big)_i * (Y_i)^{\top})$, 

где $X$ – матрица $N \times k$, $\theta$ – это $k \times m$, $Y$ – матрица $N \times m$
